```python
sns.jointplot(x="Tanggal", y="Total Penjualan", data=df, size=5)
plt.show()
sns.jointplot(x="Kode Produk", y="Penjualan Bersih", data=df, size=5)
plt.show()
```

```python
# Scatter Plot 
# x = attack, y = defense
df.plot(kind='scatter', x='Tanggal', y='Penjualan Bersih',alpha = 0.5,color = 'red')
plt.xlabel('Tanggal')              # label = name of label
plt.ylabel('Penjualan Bersih')
plt.title('Laba Bersih per tanggal')            # title = title of plot
```
